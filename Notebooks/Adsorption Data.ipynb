```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Load Data

```python
CNT65 = pd.read_excel('../Data/Adsorption/CNT Chiral 773735.xlsx')
CNT66 = pd.read_excel('../Data/Adsorption/CNT Non Chiral 775533.xlsx')
```

```python
CNT65.columns
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
plt.rcParams.update({'font.size': 16})
plt.scatter(CNT65['Relative Pressure (P/P0)'], CNT65['Quantity Absorbed (cm3/g)'], label='(6,5) CNT')
plt.scatter(CNT66['Relative Pressure (P/P0)'], CNT66['Quantity Absorbed (cm3/g)'], label='(6,6) CNT')
plt.ylabel('Quantity Absorbed ($cm^{3}$/g)')
plt.xlabel('Relative Pressure (P/P0)')
plt.legend()
plt.savefig('../Figures/Adsorption Data.png', dpi=600, bbox_inches='tight')
```
