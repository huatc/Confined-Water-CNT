```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Load Data

```python
path1 = '../Data/SANS/EQSANS_181274_merged.txt'
data1 = pd.read_csv(path1, delimiter= '\t', skiprows=2).values
path2 = '../Data/SANS/EQSANS_181275_merged.txt'
data2 = pd.read_csv(path2, delimiter= '\t', skiprows=2).values
path3 = '../Data/SANS/EQSANS_181300_merged.txt'
data3 = pd.read_csv(path3, delimiter= '\t', skiprows=2).values
path4 = '../Data/SANS/SWNT_non-chiral_8o4A_APD.txt'
data4 = pd.read_csv(path4, delimiter= '\t', skiprows=4).values
path5 = '../Data/SANS/SWNT_non-chiral_16o8A_APD.txt'
data5 = pd.read_csv(path5, delimiter= '\t', skiprows=4).values
```

```python
plt.loglog(data1[:,0], data1[:,1])
plt.loglog(data2[:,0], data2[:,1])
plt.loglog(data3[:,0], data3[:,1])
plt.loglog(data4[:,0], data4[:,1])
plt.loglog(data5[:,0], data5[:,1])
```

```python
sub_data1 = data1[:,1] - data3[:,1] 
sub_data2 = data2[:,1] - data3[:,1] 
sub_data1_error = data1[:,2] + data3[:,2]
sub_data2_error = data2[:,2] + data3[:,2]
```

```python
plt.rcParams.update({'font.size': 18})
fig, ax = plt.subplots(figsize=(7,7))
ax.errorbar(data1[:,0], sub_data1, yerr=sub_data1_error, fmt='o', color='C0', label='CNT with $D_{2}O $')
ax.errorbar(data2[:,0], sub_data2, yerr=sub_data2_error, fmt='o', color='C1', label='CNT Dry')
plt.yscale('log')
#plt.xscale('log')
plt.xlim([0, 0.6])
plt.ylim([1e-2, 1000])
plt.ylabel('I(q) (arb. unit)')
plt.xlabel('q ($\\AA^{-1}$)')
plt.legend()
#plt.savefig('../Figures/SANS_CNT.png', dpi=600, bbox_inches='tight')
```

```python
plt.rcParams.update({'font.size': 18})
plt.plot(data1[:,0], sub_data1, label='6,5 CNT with $D_{2}O $', linewidth=3)
plt.plot(data2[:,0], sub_data2, label='6,5 CNT Dry', linewidth=3)
plt.loglog(data4[:,0], data4[:,1], label= '0.84 nm')
plt.loglog(data5[:,0], data5[:,1], label= '16')
plt.yscale('log')
#plt.xscale('log')
#plt.xlim([0, 0.6])
#plt.ylim([0.01, 10])
plt.ylabel('I(q) (arb. unit)')
plt.xlabel('q ($\\AA^{-1}$)')
plt.legend()
```
